```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('TN.TXT')
```

```python
#It's all TN, so I don't need to include that data
df=df.drop('TN',axis=1)
```

```python
#Column names to make analysis possible
df.columns=['Sex','Year','Name','Occurrences']
```

```python
#How many Jessicas from my birthyear?
print(df[(df['Year'] == 2000) & df['Name'].str.findall('Jessica')])
```

```python
#I just want to work with Jessica
dfjess=df[df['Name']=='Jessica']
```

```python
dfjess['Sex'].value_counts()
```

```python
#Were any babies coded M at birth named Jessica? If so, when?
dfjess[dfjess['Sex']=='M']
```

```python
#How many total Jessicas coded M at birth?
dfjess[dfjess['Sex']=='M'].Occurrences.sum()
```

```python
dfjess.plot.scatter(x='Year',y='Occurrences',c=dfjess['Sex'].map({'F':'hotpink','M':'deepskyblue'}))
plt.title("Popularity of the name Jessica in Tennessee over time")
plt.savefig("jessica.svg")
plt.show()
#Scatter is BETTER than line because 1984, 1987 and 1989 have two occurences
```

```python
#At what height does that chart peak?
dfjess['Occurrences'].max()
```

```python
#When does that chart peak?
dfjess[dfjess['Occurrences']==1324]
```

```python
#What's the lowest?
dfjess['Occurrences'].min()
```

```python
#When was Jessica the lowest?
dfjess[dfjess['Occurrences']==5]
```
